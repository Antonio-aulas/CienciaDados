#Raspagem de Dados (webscaping)
### Este método permite capturar dados de deiversas fontes na internet: paginas html, comentários do youtube, twitter, pemitindo ao operador entender o que está sendo falando ou comentado na rede.
### Associado à análise de sentimento, pode retornar também informações se determinado produto, opinião ou fato é de caráter positivo ("gostou"), ou de carater negativo (não gostou)
Neste exemplo, vamos usar o código html de uma eleição para deputado estadual, que foi apresentada no infograficos do Estadao. Esta página não está mais disponivel na rede, e também para exemplificar o conceito de wecScraping, salvamos a página na variável html_votos

A técnica de raspagem de paginas html consiste em pesquisar tags html que formem padrões de repetição no texto, ou seja, que as informações que desejamos estejam todas designadas pela mesma tag.

```
soup = BeautifulSoup(html_votos, 'html.parser')
```

```
#varrendo o objeto soup apra pesquisar a classe custom-representative
# ao achar , le o dado da tag data-choice e insere no vetor.
votos = []
for deputado in soup.find_all("div", attrs={"class": "custom-representative"}):
  dados_deputado = []
  dados_deputado.append(deputado['data-choice']) #voto
  votos.append(dados_deputado)
```

```
import matplotlib.pyplot as plt
# criando um vetor de cores para o gráfico
cores = ['blue', 'green', 'red', 'purple', 'orange', 'cyan', 'magenta', 'yellow', 'brown', 'gray']
#criando o gráfico de barras Pessoa x Total de Votos (nome das colunas do dataframe) AO inves de usar uma cor, indicamos o vetor de cores
plt.bar(total_votos['Pessoa'], total_votos['Total de Votos'], color=cores)
# o que vai aparecer nos labels X e Y, e o título
plt.xlabel('Pessoa')
plt.ylabel('Total de Votos')
plt.title('Total de Votos por Pessoa')
#rotacionando o label X para 90 graus
plt.xticks(rotation=90)
plt.show()
```
